```python
# Generate a 12-word crypto seed phrase (BIP39 style)
import secrets
import urllib.request

# Download BIP39 English wordlist
url = 'https://raw.githubusercontent.com/bitcoin/bips/master/bip-0039/english.txt'
with urllib.request.urlopen(url) as f:
    wordlist = [line.decode('utf-8').strip() for line in f]

# Pick 12 random words
seed_phrase = ' '.join(secrets.choice(wordlist) for _ in range(12))
print(seed_phrase)
```
